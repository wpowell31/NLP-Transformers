```python
from transformers import pipeline

classifier = pipeline('text-classification')
```

```python
text = 'Hello, I am a fan of the Carnegie Mellon Tartans, Duke Blue Devils, and Kansas Jayhawks!'
```

```python
import pandas as pd

outputs = classifier(text)
pd.DataFrame(outputs)
```

```python
# Named Entity Recognition

new_tagger = pipeline('ner', aggregation_strategy='simple')

outputs = new_tagger(text)
pd.DataFrame(outputs)
```

```python
# Question Answering

reader = pipeline('question-answering')
question = 'Who is he a fan of?'
outputs = reader(question=question, context=text)
pd.DataFrame([outputs])
```

```python
# Summarization

summarizer = pipeline('summarization')
outputs = summarizer(text, min_length=10, max_length=15, clean_up_tokenization_spaces=True)
print(outputs[0]['summary_text'])
```

```python
# Text Generation

generator = pipeline('text-generation')
response = 'That is great, here are a list of some other teams.'
prompt = text + '\n\n response: \n' + response
outputs = generator(prompt, max_length=200)
outputs[0]['generated_text']
```
